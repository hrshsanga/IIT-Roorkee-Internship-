```python
# Create a dummy file for demonstration
with open("sample.txt", "w") as f:
    f.write("Hello World!\nWelcome to Python programming.\nThis is Task 1.")

# Step-by-step processing
try:
    with open("sample.txt", "r") as file:
        lines = file.readlines()  # Reads file into a list of strings (lines)
        
    line_count = len(lines)
    word_count = 0
    char_count = 0

    for line in lines:
        word_count += len(line.split()) # .split() separates words by whitespace
        char_count += len(line)         # Includes spaces and newline characters

    print(f"Lines: {line_count}")
    print(f"Words: {word_count}")
    print(f"Characters: {char_count}")

except FileNotFoundError:
    print("The specified file was not found.")
```

```python
import csv

students = [
    {"Name": "Alice", "Grade": "A", "Age": "20"},
    {"Name": "Bob", "Grade": "B", "Age": "21"},
    {"Name": "Charlie", "Grade": "A+", "Age": "19"}
]

# 1. Writing to CSV
with open("students.csv", mode="w", newline="", encoding="utf-8") as file:
    fieldnames = ["Name", "Grade", "Age"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    writer.writeheader()  # Writes the columns: Name, Grade, Age
    writer.writerows(students)
print("Data successfully written to students.csv\n")

# 2. Reading it back
print("Reading data back from CSV:")
with open("students.csv", mode="r", newline="", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(f"Student: {row['Name']} | Grade: {row['Grade']} | Age: {row['Age']}")
```

```python
def show_menu():
    print("\n--- TO-DO LIST ---")
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Exit")

def view_tasks():
    try:
        with open("todo.txt", "r") as file:
            tasks = file.readlines()
            if not tasks:
                print("\nYour to-do list is empty!")
                return
            print("\nYour Tasks:")
            for idx, task in enumerate(tasks, 1):
                print(f"{idx}. {task.strip()}")
    except FileNotFoundError:
        print("\nNo tasks found yet. Add a task to get started!")

def add_task():
    task = input("\nEnter the task: ")
    if task.strip():
        with open("todo.txt", "a") as file:  # 'a' appends to the file without overwriting
            file.write(task + "\n")
        print(f"'{task}' added successfully.")
    else:
        print("Task cannot be empty.")

# Main Application Loop
while True:
    show_menu()
    choice = input("Choose an option (1-3): ")
    
    if choice == "1":
        view_tasks()
    elif choice == "2":
        add_task()
    elif choice == "3":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please select 1, 2, or 3.")
```

```python
def safe_divide(numerator, denominator):
    try:
        result = numerator / denominator
    except ZeroDivisionError:
        print("Error: Cannot divide a number by zero.")
        return None
    except TypeError:
        print("Error: Inputs must be numerical values.")
        return None
    else:
        # Runs only if the try block executes without errors
        return result
    finally:
        # Runs regardless of what happened above
        print("Division operation attempted.")

# Test Cases
print(f"Result 1: {safe_divide(10, 2)}")   # Valid division
print(f"Result 2: {safe_divide(10, 0)}")   # Triggers ZeroDivisionError
print(f"Result 3: {safe_divide(10, 'a')}") # Triggers TypeError
```
