```python
#FizzBuzz
for i in range(1,101):
    if i%3 == 0 and i%5 == 0:
        print ("FizzBuzz")
    elif i%3 == 0:
        print('Fizz')
    elif i%5 == 0:
        print('Buzz')
    else:
        print(i)
```

```python
#Number guess game
import random
num = random.randint(1,100)
guess = 0
while guess != num:
    guess = int(input('enter num:'))
    if (guess < num):
        print('guess higher')
    elif (guess > num):
        print('guess lower:')
    else:
        print('you won')
print('draw num is:' , num)
```

```python
#pyramid star pattern
n =int(input('enter the number of rows:'))
for i in range(n-1):
    for j in range(i,n):
        print(' ', end=' ')
    for j in range (i):
        print('*', end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
```
