```python
#temperature converter C TO F & F TO C
c = float(input("enter the temp. in celsius"))
f = (c * 9/5 ) + 32
print("fahrenheit =" , f)
 
f = float(input('enter the temp. in farenheit'))
c = (f - 32) * 5/9
print("celsius =" , c)
```
