# TASKS

```python
#STUDENT GRADER
students = [("Harsh", 85),("Aman", 92),("Riya", 78)]
for name, marks in students:
    print(name, ":", marks)
```

```python
#SORT & REVERSE
names = ["Harsh", "Riya", "Aman", "Karan"]

names.sort()
print("Ascending:", names)

names.sort(reverse=True)
print("Descending:", names)
```

```python
#REMOVE DUPLICATES IN LIST
numbers = [1, 2, 3, 2, 4, 1, 5]

unique = list(set(numbers))

print(unique)
```

```python
#SQUARE 1YO 20 BY LIST COMPREHENSION
squares = [x**2 for x in range(1, 21)]

print(squares)
```

```python
#SHOPING CART
cart = []

cart.append("Milk")
cart.append("Bread")
cart.append("Eggs")

cart.remove("Bread")

print("Cart:", cart)
print("Total Items:", len(cart))
```

```python
#FIND 2ND LARGEST &SMALLEST NUMBER
nums = [10, 5, 20, 8, 15]

nums.sort()

print("2nd Smallest:", nums[1])
print("2nd Largest:", nums[-2])
```

```python
#MATRIX addition
A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

result = [[0, 0], [0, 0]]

for i in range(2):
    for j in range(2):
        result[i][j] = A[i][j] + B[i][j]

print(result)
```
