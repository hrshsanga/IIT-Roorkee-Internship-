```python
#count word frequency
def word_frequency(sentence):
    words = sentence.split()
    freq = {}

    for word in words:
        freq[word] = freq.get(word, 0) + 1

    return freq

s = input("Enter sentence: ")
print(word_frequency(s))
```

```python
#unique characters
def unique_characters(text):
    return set(text)

s = input("Enter string: ")
print(unique_characters(s))
```

```python
#word reserval 
def reverse_word(word):
    return word[::-1]

w = input("Enter word: ")
print("Reversed:", reverse_word(w))
```

```python
#palindrome checker
def palindrome(word):
    if word == word[::-1]:
        return "Yes Palindrome"
    return "Not Palindrome"

w = input("Enter word: ")
print(palindrome(w))
```

```python
#student report card
def report_card(name, marks):
    grade = "A" if marks >= 80 else "B" if marks >= 60 else "C"
    student = {"Name": name,"Marks": marks,"Grade": grade}
    return student
print(report_card("Abhi" , 89))
```

```python
#Anagram checker
def anagram(str1, str2):
    if sorted(str1) == sorted(str2):
        return "Anagram"
    return "Not Anagram"

a = input("Enter first word: ")
b = input("Enter second word: ")

print(anagram(a, b))
```

```python
#Caesar cipher encoder
def encrypt(text, shift):
    result = ""

    for ch in text:
        if ch.isalpha():
            result += chr((ord(ch) - 97 + shift) % 26 + 97)
        else:
            result += ch

    return result

print(encrypt("hello", 3))
```

```python
#Caesar cipher decoder 
def decrypt(text, shift):
    result = ""

    for ch in text:
        if ch.isalpha():
            result += chr((ord(ch) - 97 - shift) % 26 + 97)
        else:
            result += ch

    return result

print(decrypt("khoor", 3))
```
